```python
import time
affirmative = ['yes', 'yeah', 'yup', 'y', 'yea', 'mhm', 'yep', 'affirmative', 'sure', 'alright', 'okay', 'ok', 'okey-dokey', 'oky-doky', 'yessir', 'yessiree', 'okie-dokie', 'heck yeah', 'of course']
negative = ['no', 'nope', 'n', 'nah', 'nuh-uh', 'negative', 'no thanks', 'no thank you', 'heck no', 'of course not', 'never']
tasklist = []
intlo = 0

def taskask():
    msg = input("Most affirmative/negative responses are accepted.").lower()
    if msg in affirmative:
        taskcreate()
    elif msg in negative:
        print("Understood.")
        return
    else:
        print("Invalid response.")
        taskask()
def intloop(var):
    temp = 1
    try:
        if int(var) <= 0:
            return False
        temp += int(var)
        return True
    except:
        return False
def intcheck(var): #checks if the tdur input is valid
    if var == 0: #check for tdur
        global tdur
        tdur = input("Please input an integer value.")
        if intloop(tdur) == True:
            return
        else:
            print("Invalid response.")
            intcheck(0)
    elif var == 1: #check for breakintv
        global breakintv
        breakintv = input("Please input an integer value.")
        if intloop(breakintv) == True:
            return
        else:
            print("Invalid response.")
            intcheck(1)
    elif var == 2: #check for breakdur
        global breakdur
        breakdur = input("Please input an integer value.")
        if intloop(breakdur) == True:
            return
        else:
            print("Invalid response.")
            intcheck(2)
    elif var == 3: #check for notifintv
        global notifintv
        notifintv = input("Please input an integer value.")
        if intloop(notifintv) == True:
            return
        else:
            print("Invalid response.")
            intcheck(3)
    else:
        print("ERROR: UNACCEPTED PARAMETER")
        quit()
def taskcreate(): #forms the task list
    print("What is the title of your task?")
    ttitle = input("Give it a distinct title that YOU will recognize.")
    print("What is the duration of this task (in minutes)?")
    intcheck(0)
    tasklist.append({"title": ttitle, "dur": int(tdur)})
    print("Would you like to add another task?")
    taskask()
def breakcreate():
    global breaklist
    print("How often (in minutes) would you like breaks to occur?")
    print("It is advised that short breaks be taken around every 30 minutes.")
    intcheck(1)
    print("How long (in minutes) would you like breaks to last?")
    print("(Breaks around 10 minutes long are good for productivity.)")
    intcheck(2)
    breaklist = [int(breakintv), int(breakdur)]
def notifcreate():
    print("How often (in minutes) would you like to receive session notifications?")
    print("(Allows users with hidden timers to keep track of their session duration.)")
    intcheck(3)
def singularitycheck(term):
    global ess
    if int(term) == 1:
            ess = "."
    else:
        ess = "s."

taskcreate()
print(tasklist)
breakcreate()
print(breaklist)
notifcreate()
print(notifintv)
singularitycheck(breakintv)
breaksummary = "You have chosen to take " + breakdur + "-minute breaks every " + breakintv + " minute" + ess
singularitycheck(notifintv)
notifsummary = "You have also chosen to receive session notifications every " + notifintv + " minute" + ess
def tasksummary():
    global intlo
    if intlo < len(tasklist):
        thistask = tasklist[intlo]
        print("Assignment Title: " + thistask["title"])
        print("Expected Duration: " + str(thistask["dur"]) + '\n')
        intlo += 1
        tasksummary()
    else:
        intlo = 0
        return
print(breaksummary + "\n" + notifsummary)
print("You have input these tasks to complete during your work session:\n")
tasksummary()
print("All done.")
```

```python
secs = -1
mins = 0
tasknum = 0
watchon = True

def notifcheck():
    if mins == 0:
        return
    if (mins % int(notifintv)) == 0:
        singularitycheck(mins)
        print("The timer has been running for more than " + str(int(mins)) + " minute" + ess)
    else:
        return
def breakcheck():
    global breaktime
    if mins == 0:
        return
    if (mins % int(breakintv)) == 0:
        print("It's break time! Please take a " + breakdur + "-minute break.")
        breaktime = True
    else:
        return
def taskcheck(tasknum):
    if mins == 0:
        return
    if tasknum < len(tasklist):
        currentt = tasklist[tasknum]
        if (mins % int(currentt["dur"])) == 0:
            print('You anticipated that the task "' + currentt["title"] + '" would be done by now.')
        else:
            return
    else:
        watchon = False
def inctime():
    global secs
    global mins
    secs += 1
    mins = secs / 60
    dissecs = secs - int(secs / 60) * 60
    dismins = int(mins - int(mins / 60) * 60)
    notifcheck() #Calculates if a notification should be provided
    breakcheck() #Calculates if it's break time
    #taskcheck() #Calculates if the task was expected to have been completed
    print(str(dismins) + "m : " + str(dissecs) + "s                     ", end = '\r')
    time.sleep(1)
    inctime()

while watchon == True:
    inctime()
if watchon == False:
    print("Your work session lasted " + str(mins) + " minutes and " + str(secs) + " seconds.")
```

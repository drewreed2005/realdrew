Below is a placeholder process of saving user information. Here are some important frontend considerations:
- `taskask()` should be in the form of two text inputs: the left (labeled "Task Title") allowing the user to input a 50-character long task name and the right (labeled "Task Duration") exclusively allowing the user to input a three-digit integer
- `breakdur`, `breakintv`, and `notifintv` will all be defined by a text input box identical to the right one of `taskask()`
- The task settings should be a separate section of the menus, but the notifications and break settings can all be in one place.

```python
import time
affirmative = ['yes', 'yeah', 'yup', 'y', 'yea', 'mhm', 'yep', 'affirmative', 'sure', 'alright', 'okay', 'ok', 'okey-dokey', 'oky-doky', 'yessir', 'yessiree', 'okie-dokie', 'heck yeah', 'of course']
negative = ['no', 'nope', 'n', 'nah', 'nuh-uh', 'negative', 'no thanks', 'no thank you', 'heck no', 'of course not', 'never'] #just for this beta program's simplicity
tasklist = [] #to be filled out as an appended list of dictionaries
intlo = 0 #a global variable that always resets to zero after use

def taskask():
    msg = input("Most affirmative/negative responses are accepted.").lower() #checks for a second task
    if msg in affirmative:
        taskcreate()
    elif msg in negative:
        print("Understood.")
        return
    else:
        print("Invalid response.")
        taskask()
def intloop(var): #checks if a given parameter input is an integer greater than 0
    temp = 1
    try:
        if int(var) <= 0:
            return False
        temp += int(var)
        return True
    except:
        return False
def intcheck(var): #checks if the var input is valid, then defines it (can be iterated)
    if var == 0: #check for tdur
        global tdur
        tdur = input("Please input an integer value.")
        if intloop(tdur) == True:
            return
        else:
            print("Invalid response.")
            intcheck(0)
    elif var == 1: #check for breakintv
        global breakintv
        breakintv = input("Please input an integer value.")
        if intloop(breakintv) == True:
            return
        else:
            print("Invalid response.")
            intcheck(1)
    elif var == 2: #check for breakdur
        global breakdur
        breakdur = input("Please input an integer value.")
        if intloop(breakdur) == True:
            return
        else:
            print("Invalid response.")
            intcheck(2)
    elif var == 3: #check for notifintv
        global notifintv
        notifintv = input("Please input an integer value.")
        if intloop(notifintv) == True:
            return
        else:
            print("Invalid response.")
            intcheck(3)
    else:
        print("ERROR: UNACCEPTED PARAMETER")
        quit()
def taskcreate(): #forms the task list
    print("What is the title of your task?")
    ttitle = input("Give it a distinct title that YOU will recognize.")
    print("What is the duration of this task (in minutes)?")
    intcheck(0)
    tasklist.append({"title": ttitle, "dur": int(tdur)})
    print("Would you like to add another task?")
    taskask()
def breakcreate():
    global breaklist
    print("How often (in minutes) would you like breaks to occur?")
    print("It is advised that short breaks be taken around every 30 minutes.")
    intcheck(1)
    print("How long (in minutes) would you like breaks to last?")
    print("(Breaks around 10 minutes long are good for productivity.)")
    intcheck(2)
    breaklist = [int(breakintv), int(breakdur)]
def notifcreate():
    print("How often (in minutes) would you like to receive session notifications?")
    print("(Allows users with hidden timers to keep track of their session duration.)")
    intcheck(3)
def singularitycheck(term): #checks if an inputted number is singular (used for grammar)
    global ess
    if int(term) == 1:
            ess = "."
    else:
        ess = "s."

taskcreate()
print(tasklist) #debug check for tasklist formatting
breakcreate()
print(breaklist) #debug check for established break time info
notifcreate()
print(notifintv)
singularitycheck(breakintv)
breaksummary = "You have chosen to take " + breakdur + "-minute breaks every " + breakintv + " minute" + ess
singularitycheck(notifintv)
notifsummary = "You have also chosen to receive session notifications every " + notifintv + " minute" + ess
def tasksummary():
    global intlo
    if intlo < len(tasklist):
        thistask = tasklist[intlo]
        print("Assignment Title: " + thistask["title"])
        print("Expected Duration: " + str(thistask["dur"]) + ' minutes \n')
        intlo += 1
        tasksummary()
    else:
        intlo = 0
        return
print(breaksummary + "\n" + notifsummary)
print("You have input these tasks to complete during your work session:\n")
tasksummary()
print("All done.")
```

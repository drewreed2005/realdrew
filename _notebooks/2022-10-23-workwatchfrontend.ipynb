Millisecond-based timer.

```python
let output = document.getElementById('stopwatch');
let ms = 0;
let sec = 0;
let min = 0;
let running = 0;

function timer() {
    ms++;
    if(ms >= 100){
        sec++
        ms = 0
    }
    if(sec === 60){
        min++
        sec = 0
    }
    if(min === 60){
        ms, sec, min = 0;
    }

    //Doing some string interpolation
    let milli = ms < 10 ? `0`+ ms : ms;
    let seconds = sec < 10 ? `0`+ sec : sec;
    let minute = min < 10 ? `0` + min : min;

    let timer= `${minute}:${seconds}:${milli}`;
    output.innerHTML =timer;
};
//Start timer
function start(){
 if (running == 0){
   time = setInterval(timer,10)
   running = 1
 }
}
//stop timer
function stop(){
    running = 0
    clearInterval(time)
}
//reset timer
function reset(){
    running = 0
    ms = 0;
    sec = 0;
    min = 0;

    output.innerHTML = `00:00:00`
}
const startBtn = document.getElementById('startBtn');
const stopBtn =  document.getElementById('stopBtn');
const resetBtn = document.getElementById('resetBtn');

startBtn.onclick = function() {
  start()
};
stopBtn.onclick = function() {
  stop()
};
resetBtn.onclick = function() {
  reset()
};
```

Second-based timer (with hour count).

```python
let output = document.getElementById('stopwatch');
let constant = 0;
let sec = 0;
let min = 0;
let hr = 0;
let running = 0;

function timer() {
    constant++;
    sec++;
    if(sec === 60){
        min++
        sec = 0
    }
    if(min === 60){
        hr++
        min = 0
    }

    //Doing some string interpolation
    let seconds = sec < 10 ? `0`+ sec : sec;
    let minute = min < 10 ? `0` + min : min;
    let hour = hr < 10 ? `0` + hr : hr;

    let timer= `${hour}:${minute}:${seconds}`;
    output.innerHTML =timer;
  }
//Start timer
function start(){
 if (running == 0){
   time = setInterval(timer,1000)
   running = 1
 }
}
//stop timer
function stop(){
    running = 0
    clearInterval(timer)
}
//reset timer
function reset(){
    running = 0;
    sec = 0;
    min = 0;
    hr = 0;
    output.innerHTML = `00:00:00`
}
const startBtn = document.getElementById('startBtn');
const stopBtn =  document.getElementById('stopBtn');
const resetBtn = document.getElementById('resetBtn');

startBtn.onclick = function() {
  start()
};
stopBtn.onclick = function() {
  stop()
};
resetBtn.onclick = function() {
  reset()
};
```

```python
let output = document.getElementById('stopwatch');
let constant = 0; //Below are timer run var
let ms = 0;
let sec = 0;
let min = 0;
let hr = 0;
let running = 0;

function timer() {
    constant++;
    ms++;
    if(ms >= 100){
        sec++
        ms = 0
    }
    if(sec === 60){
        min++
        sec = 0
    }
    if(min === 60){
        hr++
        min = 0
    }

    //Doing some string interpolation
    let seconds = sec < 10 ? `0`+ sec : sec;
    let minute = min < 10 ? `0` + min : min;
    let hour = hr < 10 ? `0` + hr : hr;

    let time = `${hour}:${minute}:${seconds}`;
    output.innerHTML =time;
  }
//Start timer
function start(){
 if (running == 0){
   time = setInterval(timer,10)
   running = 1
 }
}
//stop timer
function stop(){
    running = 0
    clearInterval(time)
}
//reset timer
function reset(){
    stop()
    running = 0;
    sec = 0;
    min = 0;
    hr = 0;
    output.innerHTML = `00:00:00`
}
const startBtn = document.getElementById('startBtn');
const stopBtn =  document.getElementById('stopBtn');
const resetBtn = document.getElementById('resetBtn');

startBtn.onclick = function() {
  start()
};
stopBtn.onclick = function() {
  stop()
};
resetBtn.onclick = function() {
  reset()
}

var hidden = 0;
var hw = document.getElementById('hide');
var sw = document.getElementById('stopwatch');
hw.onclick = function() {
    if (hidden == 0) {
      sw.style.display = "none"
      hidden = 1
    } else {
      sw.style.display = "block"
      hidden = 0
    };
  };
```
